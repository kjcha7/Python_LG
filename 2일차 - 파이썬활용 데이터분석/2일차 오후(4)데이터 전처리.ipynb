### 1. 누락된 값 골라내기

- 누락된 데이터를 골라내는 방법
- dropna

```python
from numpy import nan as NA
data = pd.Series([1, NA, 3.5, NA, 7])
data.dropna()
```

- 불리언 색인을 이용해서 직접 계산하는 것도 가능

```python
data[data.notnull()]
```

- DataFrame의 경우, 모두 NA인 행이나 컬럼을 제외하든가, 하나라도 NA인 값을 포함하고 있는 행이나 컬럼을 제외시킬수도 있음
- Dropna : 기본적으로 NA값이 하나라도 있는 행은 제외시킨다. 

```python
data = pd.DataFrame([[1., 6.5, 3.], [1., NA, NA],
                     [NA, NA, NA], [NA, 6.5, 3.]])

data
```

```python
cleaned = data.dropna()
cleaned
```

- how = 'all'옵션을 주면 모든값이 NA인 행만 제외시킨다. 

```python
data.dropna(how='all')
```

- 컬럼을 제외시키는 방법은 옵션으로 axis=1을 주면 행을 제외시키는것과 동일한 방식으로 동작한다. 

```python
data[4] = NA
data
```

```python
data.dropna(axis=1, how='all')
```

- DataFrame의 행을 제외시키는 방법은 주로 시계열 데이터에 사용됨
- 몇개이상의 값이 들어있는 행만 살펴보고 싶다면? thresh인자에 원하는 값을 넘기면 됨

```python
df = pd.DataFrame(np.random.randn(7, 3))
df.iloc[:4, 1] = NA
df.iloc[:2, 2] = NA
df
```

```python
df.dropna()
```

```python
df.dropna(thresh=2)
```

### 2. 함수나 매핑 이용해 데이터 변형하기

```python
data = pd.DataFrame({'food': ['bacon', 'pulled pork', 'bacon',
                              'Pastrami', 'corned beef', 'Bacon',
                              'pastrami', 'honey ham', 'nova lox'],
                     'ounces': [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data
```

- 해당 육류가 어떤 동물의 고기인지 알려줄 수 있는 칼럼을 하나 추가한다고 가정하고, 육류별 동물을 담고 있는 사전 데이터를 만들자

```python
meat_to_animal = {
  'bacon': 'pig',
  'pulled pork': 'pig',
  'pastrami': 'cow',
  'corned beef': 'cow',
  'honey ham': 'pig',
  'nova lox': 'salmon'
}
```

- 이 데이터에는 육류의 이름에 대.소문자가 섞여 있는 사소한 문제가 있음으로 모두 변경해야함

```python
lowercased = data['food'].str.lower()
lowercased
```

- Series의 map메서드는 사전류의 객체나 어떤 함수를 받을 수 있음

```python
data['animal'] = lowercased.map(meat_to_animal)
data
```

### 3. 값 치환하기(replace)

```python
data = pd.Series([1., -999., 2., -999., -1000., 3.])
data
```

```python
data.replace(-999, np.nan)
```

- 여러개의 값을 한번에 치환하려면 하나의 값 대신 치환하려는 값의 리스트를 넘기면 됨

```python
data.replace([-999, -1000], np.nan)
```

- 치환하려는 값마다 다른 값으로 치환하려면 누락된 값 대신 새로이 지정할 값의 리스트를 사용해야함

```python
data.replace([-999, -1000], [np.nan, 0])
```

- 2개의 리스트 대신 사전을 이용해도 됨

```python
data.replace({-999: np.nan, -1000: 0})
```

### 4. 축 색인 이름 바꾸기

- Series의 값처럼 축 이름 역시 유사한 방식으로 함수나 새롭게 바꿀 값으로 이용해서 변형할 수 있음
- 특히, 새로운 자료구조를 만들지 않고 그 자리에서 바로 축 이름을 변경하는것이 가능함!

```python
data = pd.DataFrame(np.arange(12).reshape((3, 4)),
                    index=['Ohio', 'Colorado', 'New York'],
                    columns=['one', 'two', 'three', 'four'])
```

```python
data
```

```python
x = lambda a : a + 10
print(x(5))
```

```python
transform = lambda x: x.upper()
transform
```

- 대문자로 변경된 축 이름을 DataFrame의 index에 바로 대입할 수 있음

```python
data.index.map(transform)
```

```python
data.index = data.index.map(transform)
data
```

    - 원래 객체를 변경하지 않고 새로운 객체를 생성하려면 rename 메서드를 사용한다

```python
data.rename(index=str.title, columns=str.upper)
```

    - 특히 rename메서드는 사전 형식의 객체를 이용해서 축 이름 중 일부만 변경하는 것도 가능하다

```python
data.rename(index={'OHIO': 'INDIANA'},
            columns={'three': 'peekaboo'}, inplace=True)
```

- rename메서드를 사용하면 DataFrame을 직접 복사해서 index와 columns속성을 갱신할 필요없이 바로 변경할 수 있음
- 원본 데이터를 바로 변경하려면 inplace=True옵션을 넘겨주면 됨

```python
data.rename(index={'OHIO': 'INDIANA'})
data
```

### 5. 개별화와 양자화

- 연속성 데이터는 종종 개별로 분할하거나 분석을 위해 그룹별로 나누기도 함
- 수업에 참여하는 학생 그룹데이터가 있고, 나이대에 따라 분류한다고 가정해보자

```python
ages = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]
```

- 이 데이터를 pandas의 cut함수를 사용해서 18-25, 26-35, 35-60, 60이상인 그룹으로 나누어 보자

```python
bins = [18, 25, 35, 60, 100]
cats = pd.cut(ages, bins)
cats
```

- pandas에서 반환하는 객체는 Categorical객체라는 특수한 객체인데, 이 객체는 그룹 이름이 담긴 배열이라고 생각하면 됨
- Categorical객체는 codes속성에 있는 ages데이터에 대한 카테고리 이름을 categories라는 배열에 내부적으로 담고 있다. 

```python
cats.codes
```

```python
cats.categories
```

```python
pd.value_counts(cats)
```

- 간격을 나타내는 표기법은 괄호로 시작해서 대괄호로 끝나는데, 괄호쪽의 값은 포함하지 ㅇ낳고, 대관로 쪽의 값은 포함하는 간격을 나타냄
- right=False를 넘겨서 괄호와 대괄호의 위치를 바꿀수도 있음

```python
pd.cut(ages, [18, 26, 36, 61, 100], right=False)
```

- labels옵션으로 그룹의 이름을 직접 넘겨줄 수도 있음

```python
group_names = ['Youth', 'YoungAdult', 'MiddleAged', 'Senior']
pd.cut(ages, bins, labels=group_names)
```

- 만약, cut함수에 명시적으로 그룹의 경계값을 넘기지 않고 그룹의 개수를 넘겨주면 데이터에서 최소값과 최대값을 기준으로 균등한 길이의 그룹을 자동으로 계산함

- 어떤 균등분포내에서 4개의 그룹으로 나눠보자

```python
data = np.random.rand(20)
pd.cut(data, 4, precision=2)
```

- 이를 위한 가장 적합한 함수로 qcut이 있는데, 이는 표본 변위치를 기반으로 데이터를 나눠준다
- cut함수를 사용하게 되면 데이터의 분산에 따라 각각의 그룹마다 데이터의 개수가 다르게 나누어 지는 경우가 많다
- 반면, qcut은 표준 변위치를 사용하기 때문에 적당히 같은 크기의 그룹으로 나눌수 있다. 

```python
data = np.random.randn(1000)  # Normally distributed
cats = pd.qcut(data, 4)  # Cut into quartiles
cats
```

```python
pd.value_counts(cats)
```

- cut함수와 유사하게 변위치를 직접 지정해 줄 수 있다. 

```python
pd.qcut(data, [0, 0.1, 0.5, 0.9, 1.])
```

-  그룹분석과 변위치를 다룰 때는 cut과 qcut함수 같은 이산 함수가 특히 더 유용함!

### 6. 특이값(Outliers) 찾아내고 제거하기

- 배열연산을 수행할 때는 특이값(Outliers)를 제외하거나 적당한 값으로 대체하는 것이 중요함. 

```python
data = pd.DataFrame(np.random.randn(1000, 4))
data.describe()
```

- 이 DataFrame의 한 컬럼에서 절대 값이 3을 초과하는 값을 찾아내자!

```python
col = data[2]
col[np.abs(col) > 3]
```

- 절대 값3을 초과하는 값이 들어있는 모든 행을 선택하려면 불리언 DataFrame에서 any메서드를 사용하면 됨

```python
data[(np.abs(data) > 3).any(1)]
```

### 7. 표시자/더미변수

- 통계모델이나 기계학습을 위한 필수 데이터 변형은 분류 값을 더미나 표시행렬로 반환하는 것임!
- 만약, 어떤 DataFrame의 한 컬럼에 k가지의 값이 있다면, k개의 칼럼이 있는 DataFrame이나 행렬를 만들고 값으로는 1, 0을 채워 넣을 것이다. 
- 이때 사용하는 함수 -> get_dummies함수

```python
df = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'b'],
                   'data1': range(6)})
df
```

```python
pd.get_dummies(df['key'])
```

- 컬럼이름에 접두어를 추가한후 다른 데이터와 병합하고 싶은때는 prefix인자를 사용하면 됨

```python
dummies = pd.get_dummies(df['key'], prefix='key')
df_with_dummy = df[['data1']].join(dummies)
df_with_dummy
```

- DataFrame의 한 칼럼이 여러 카테고리에 속한다면?
- 각 장르마다 표시용 값을 추가하려면 먼저 데이터 묶음에서 유일한 장르 목록부터 추출해야함

```python
mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('examples/movies.dat', sep='::',
                       header=None, names=mnames)
movies[:10]
```

```python
all_genres = []
for x in movies.genres:
    all_genres.extend(x.split('|'))
genres = pd.unique(all_genres)
```

```python
genres
```

- 이제 표시용 DataFrame을 생성하기 위해 0으로 초기화된 DataFrame을 생성하자

```python
zero_matrix = np.zeros((len(movies), len(genres)))
dummies = pd.DataFrame(zero_matrix, columns=genres)
dummies
```

```python
gen = movies.genres[0]
gen.split('|')
dummies.columns.get_indexer(gen.split('|'))
```

- 각 영화를 순회하면서 dummies칼럼의 각 항목을 1로 설정한다. 

```python
for i, gen in enumerate(movies.genres):
    indices = dummies.columns.get_indexer(gen.split('|'))
    dummies.iloc[i, indices] = 1
```

- 그런다음, movies DataFrame과 조합하면 됨

```python
movies_windic = movies.join(dummies.add_prefix('Genre_'))
movies_windic.iloc[0]
```
