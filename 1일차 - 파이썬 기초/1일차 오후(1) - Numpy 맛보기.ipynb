- 배열명[행위치, 열위치] 
-- #행위치만 입력하면 지정한 행 전체가 선택됨

````python
a2= np.arange(10, 100, 10).reshape(3,3)
a2
````

````python
#행위지가 0이고 열위치가 2인 원소를 선택해서 가져오기
a2[0,2]
````

````python
#2차원배열의 행과 열의 위치를 지정해서 원소를 선택한 후 값을 변경
a2[2,2]=95
a2
````

````python
#행위치 전체를 지정해 행 전체를 가져오기
a2[1]
````

````python
#행전체를 변경하려면?
a2[1] = np.array([45,55,65])
````

````python
a2
````

````python
#2차원배열에서 행과 열의 위치를 지정해 여러 원소를 선택해서 가져오는 예
a2[[0,2], [0,1]]
````

````python
#배열에 조건을 지정해 조건을 만족하는 배열을 선택
a = np.array([1,2,3,4,5,6,7,8,])
a
````

````python
a[a>3]
````

````python
#a열에서 짝수 원소만 선택하려면?
a[a%2 == 0]
````

- 배열의 슬라이싱

````python
#배열[시작위치:끝위치]
b1=np.array([0,1,2,3,4,5])
b1[1:4]
````

````python
b1[:3]
````

````python
b1[2:]
````

- 원소 변경하기

````python
b1[2:5] = np.array([10,11,12])
````

````python
b1
````

````python
#여러원소의 값을 같은 값으로 변경하려면?
b1[3:6]=60
````

````python
b1
````

````python
#특정행을 선택한후 열을 슬라이싱하려면?
#배열[행위치][열의시작위치:열의끝위치]
````

````python
b2=np.arange(10, 100, 10).reshape(3,3)
b2
````

````python
b2[1:3, 1:3]
````

````python
b2[:3, 1:]
````

````python
#2차원배열에서 행을 지정하고 열을 슬라이싱하는 예
b2[1][0:2]
````

````python
b2[0:2, 1:3] = np.array([[25,35], [55, 65]])
````

````python
b2
````
